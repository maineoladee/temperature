### Data Preprocessing

Removing unstructured part of data: <br>
Farenheit included in every row of temperatures will be removed, only celcius data is required for this analysis

```python
# Creating a class and functions to import data, get_data and remove ...,
# fahrenheit equivalents of celsius recorded in data using regex
# rearrange columns strings LHS, float RHS,
# cast temperature in Jan:Year to float from strings
class TemperatureData:
    def __init__(self, filename):
        self.df = pd.read_csv(filename)

    def get_data(self):
        return self.df

    def remove_fahrenheit(self):
        pattern_1 = r"\([^)]*\)"
        pattern_2 = "\n"
        for col in self.df.columns:
            self.df[col] = self.df[col].apply(lambda x: re.sub(pattern_1, "", x))
            self.df[col] = self.df[col].apply(lambda x: re.sub(pattern_2, "", x))
            
    def rearrange_columns(self):
        cols = ['Country', 'City', 'Continent', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec', 'Year']
        self.df = self.df[cols]
            
    def cast_to_float(self):
        for col in self.df.columns[3:]:
            self.df[col] = self.df[col].astype(float)

# calling class, functions and methods to read  in data
data = TemperatureData("Average Temperature of Cities.csv")
data.get_data()
```

```python
# function call to remove fahrenheit, 
# rearrange columns in order of objects and numeric
# Cast temperatures from strings to float
data.remove_fahrenheit()
data.rearrange_columns()
data.cast_to_float()
```

```python
# Extracting series needed for analysis from dataset
cold_data = dataset[["Country","City","Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec",]]
```

```python
cold_data
```

```python
# Function shows each country, city, month that has the coldest temperature recorded 

# group by country
grouped = cold_data.groupby("Country")

# loop through each group
for name, group in grouped:
    # get the city and month where the coldest temperature was recorded
    city = (
        group[
            [
                "City",
                "Jan",
                "Feb",
                "Mar",
                "Apr",
                "May",
                "Jun",
                "Jul",
                "Aug",
                "Sep",
                "Oct",
                "Nov",
                "Dec",
            ]
        ]
        .melt(id_vars=["City"], var_name="Month", value_name="Temperature")
        .sort_values(by="Temperature")
        .iloc[0]
    )
    # print the result
    print(
        name
        + ": "
        + city["City"]
        + " - "
        + city["Month"]
        + " ("
        + str(city["Temperature"])
        + "°C)"
    )
```
